1. 1. What is the difference between a function and a method in Python?

Ans-->

*   Function: A block of code that is defined using the def keyword and can be called independently.

6. Explain the concept of generators in Python and how they are defined.

Ans-->

*   Generators produce items one at a time using the yield keyword.


*   They do not store the entire sequence in memory.



```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)
```

7. What are the advantages of using generators over regular functions?

Ans-->

*   Memory Efficient: Generates one item at a time.
*   Faster for large datasets.

*   Simpler syntax than using classes for iterators.





```python
# Regular list returns all elements at once
def get_numbers():
    return [i for i in range(1000000)]

# Generator yields one by one
def generate_numbers():
    for i in range(1000000):
        yield i
```

4.  Check if a number is prime (1 to 200)

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Example usage:
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
```

5. Fibonacci sequence iterator class

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.n1, self.n2 = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.max_terms:
            result = self.n1
            self.n1, self.n2 = self.n2, self.n1 + self.n2
            self.count += 1
            return result
        else:
            raise StopIteration
```

6.  Generator function for powers of 2

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

7.  Generator function to read file line by line



```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

10. Remove all vowels from a string using filter()

```python
def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

# Example
result = remove_vowels("Hello World")
```

11. Bookshop Accounting Routine using Lambda and Map

```python
orders = [
    ["Order001", "Books", 45.50, 1],
    ["Order002", "Notebooks", 10.00, 5],
    ["Order003", "Pen", 1.50, 60],
    ["Order004", "Stationery", 150.00, 1]
]

# Lambda with map to calculate total and add â‚¬10 if total < 100
final_bill = list(map(
    lambda x: (x[0], x[2]*x[3] + 10 if x[2]*x[3] < 100 else x[2]*x[3]),
    orders
))

# Output the result
print(final_bill)
```
